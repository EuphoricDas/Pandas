# Pandas Visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\bindu\Documents\DA\Python\Pandas\Ice Cream Ratings.csv")
```

```python
df
```

```python
df.set_index('Date')
```

```python
# Different types of styling that could be done to the visualisations
print(plt.style.available)
```

```python
plt.style.use('fivethirtyeight')
```

```python
df.plot()
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'line', subplots=True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel='Daily Rating', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
#Visualise just one column
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
#horizontal stacked bar chart
df.plot.barh(stacked = True)
```

```python
#Scatter plot
# s-size of the dot. c- color of the dot
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Purple')
```

```python
#Histogram- generally used to see the distribution of variables
df.plot.hist()
```

```python
df.plot.hist(bins = 20)
```

```python
#BoxPlot- We can get the min, 25, 50, 75 percentile and the max and how
# it is skewed
df.boxplot()
```

```python
# Area plot
df.plot.area(figsize = (10,5))
```

```python
# Pie Plot
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```
