Are there certain continents that have grown faster and if so, by how much?

```python
df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by='2022 Population', ascending = False)
```

```python
# Lets visualise it
df1 = df.groupby('Continent')['1970 Population','1980 Population','1990 Population','2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population'].mean().sort_values(by='2022 Population', ascending = False)
df1.plot()
```

```python
# Now lets transpose the df to make the population as rows and the 
# continents as indexes or columns just for a better visualisation
df2 = df1.transpose()
df2
```

```python
df2.plot()
```
